Now we start implementing gradient descent in plain python. Again the goal is to come up with same w1, w2 and bias that keras model calculated. We want to show how keras/tensorflow would have computed these values internally using gradient descent
First write couple of helper routines such as sigmoid and log_loss

```python
def sigmoid_numpy(X):
   return 1/(1+np.exp(-X))
```

```python
def log_loss(y_true, y_predicted):
    epsilon = 1e-15
    y_predicted_new = [max(i,epsilon) for i in y_predicted]
    y_predicted_new = [min(i,1-epsilon) for i in y_predicted_new]
    y_predicted_new = np.array(y_predicted_new)
    return -np.mean(y_true*np.log(y_predicted_new)+(1-y_true)*np.log(1-y_predicted_new))
```

```python
def gradient_descent(Pregnancies, BloodPressure, SkinThickness, BMI,
       DiabetesPedigreeFunction, Age, y_true, epochs, loss_thresold):
    w1 = w2 = w3 = w4 = w5 = w6 = 1
    bias = 0
    rate = 0.5
    n = len(Pregnancies)
    for i in range(epochs):
        weighted_sum = w1 * Pregnancies + w2 * BloodPressure + w3 * SkinThickness + w4 * BMI + w5 * DiabetesPedigreeFunction + w6 * Age +bias
        y_predicted = sigmoid_numpy(weighted_sum)
        loss = log_loss(y_true, y_predicted)

        w1d = (1/n)*np.dot(np.transpose(Pregnancies),(y_predicted-y_true))
        w2d = (1/n)*np.dot(np.transpose(BloodPressure),(y_predicted-y_true))
        w3d = (1/n)*np.dot(np.transpose(SkinThickness),(y_predicted-y_true))
        w4d = (1/n)*np.dot(np.transpose(BMI),(y_predicted-y_true))
        w5d = (1/n)*np.dot(np.transpose(DiabetesPedigreeFunction),(y_predicted-y_true))
        w6d = (1/n)*np.dot(np.transpose(Age),(y_predicted-y_true))

        bias_d = np.mean(y_predicted - y_true)
        w1 = w1 - rate * w1d
        w2 = w2 - rate * w2d
        w3 = w3 - rate * w3d
        w4 = w4 - rate * w4d
        w5 = w5 - rate * w5d
        w6 = w6 - rate * w6d
        bias = bias - rate * bias_d

        print (f'Epoch:{i}, w1:{w1}, w2:{w2}, w3:{w3}, w4:{w4}, w5:{w5}, w6:{w6}, bias:{bias}, loss:{loss}')

        if loss<=loss_thresold:
            break

    return w1, w2, w3, w4, w5, w6, bias
```

**Now we will create a class NN which has this gradient descent function which we will use to train our model**

```python
class myNeuralNetwork():
  def __init__(self):
    self.w1 = 1,
    self.w2 = 1,
    self.w3 = 1,
    self.w4 = 1,
    self.w5 = 1,
    self.w6 = 1,
    self.bias = 0

  def fit(self, X, y, epochs, loss_threshold):
    self.w1, self.w2, self.w3, self.w4, self.w5, self.w6, self.bias = self.gradient_descent(X['Pregnancies'],X['BloodPressure'],X['SkinThickness'],X['BMI'],X['DiabetesPedigreeFunction'],X['Age'], y, epochs, loss_threshold)

  def predict(self, X_test):
    weighted_sum =  self.w1 * X_test['Pregnancies'] + self.w2 * X_test['BloodPressure'] + self.w3 * X_test['SkinThickness'] + self.w4 * X_test['BMI']+ self.w5 * X_test['DiabetesPedigreeFunction']+ self.w6 * X_test['Age'] + self.bias
    return sigmoid_numpy(weighted_sum)

  def gradient_descent(self, Pregnancies, BloodPressure, SkinThickness, BMI,
       DiabetesPedigreeFunction, Age, y_true, epochs, loss_thresold):
    w1 = w2 = w3 = w4 = w5 = w6 = 1
    bias = 0
    rate = 0.5
    n = len(Pregnancies)
    for i in range(epochs):
        weighted_sum = w1 * Pregnancies + w2 * BloodPressure + w3 * SkinThickness + w4 * BMI + w5 * DiabetesPedigreeFunction + w6 * Age +bias
        y_predicted = sigmoid_numpy(weighted_sum)
        loss = log_loss(y_true, y_predicted)

        w1d = (1/n)*np.dot(np.transpose(Pregnancies),(y_predicted-y_true))
        w2d = (1/n)*np.dot(np.transpose(BloodPressure),(y_predicted-y_true))
        w3d = (1/n)*np.dot(np.transpose(SkinThickness),(y_predicted-y_true))
        w4d = (1/n)*np.dot(np.transpose(BMI),(y_predicted-y_true))
        w5d = (1/n)*np.dot(np.transpose(DiabetesPedigreeFunction),(y_predicted-y_true))
        w6d = (1/n)*np.dot(np.transpose(Age),(y_predicted-y_true))

        bias_d = np.mean(y_predicted - y_true)
        w1 = w1 - rate * w1d
        w2 = w2 - rate * w2d
        w3 = w3 - rate * w3d
        w4 = w4 - rate * w4d
        w5 = w5 - rate * w5d
        w6 = w6 - rate * w6d
        bias = bias - rate * bias_d

        if i%50 == 0:
          print (f'Epoch:{i}, w1:{w1}, w2:{w2}, w3:{w3}, w4:{w4}, w5:{w5}, w6:{w6}, bias:{bias}, loss:{loss}')

        if loss<=loss_thresold:
          print (f'Epoch:{i}, w1:{w1}, w2:{w2}, w3:{w3}, w4:{w4}, w5:{w5}, w6:{w6}, bias:{bias}, loss:{loss}')
          break

    return w1, w2, w3, w4, w5, w6, bias
```
